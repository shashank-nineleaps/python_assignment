```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
movies_data = './Datasets/movies_data_cleaned/combined_dataset.csv'
```

```python
movies_df = pd.read_csv(movies_data)
```

```python
movies_df = movies_df.drop_duplicates(subset=['Title'], keep='first')
```

```python
movies_df = movies_df.drop_duplicates(subset=['MovieID'], keep='first')
```

```python
movies_df = pd.read_csv(movies_data)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
```

```python
genre_boxoffice = movies_df.groupby('Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False).reset_index()
```

```python
sns.barplot(data=genre_boxoffice, x='BoxOffice (Crores)', y='Genre', hue='Genre', palette='coolwarm', legend=False)
plt.title('Average BoxOffice Collection by Genre (All Industries)')
plt.xlabel('Average BoxOffice (Crores)')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
bollywood_only = movies_df[movies_df['Industry'] == 'Bollywood']

plt.figure(figsize=(14, 8))
bolly_genre_boxoffice = bollywood_only.groupby('Genre')['BoxOffice (Crores)'].mean().sort_values(ascending=False).reset_index()

sns.barplot(data=bolly_genre_boxoffice, x='BoxOffice (Crores)', y='Genre', hue='Genre', palette='crest')
plt.title('Average BoxOffice Collection by Genre (Bollywood Only)')
plt.xlabel('Average BoxOffice (Crores)')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
