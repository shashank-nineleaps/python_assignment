```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))
```

```python
movies_data = './Datasets/movies_data_cleaned/combined_dataset.csv'
```

```python
movies_df = pd.read_csv(movies_data)
```

```python
plt.subplot(2, 1, 1)
sns.histplot(movies_df['BoxOffice (Crores)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of BoxOffice Collections')
plt.xlabel('BoxOffice (Crores)')
plt.ylabel('Frequency')
```

```python
plt.subplot(2, 1, 2)
sns.boxplot(x=movies_df['BoxOffice (Crores)'], color='orange')
plt.title('Boxplot of BoxOffice Collections')
plt.xlabel('BoxOffice (Crores)')
```

```python
plt.tight_layout()
plt.show()
```

```python
Q1 = movies_df['BoxOffice (Crores)'].quantile(0.25)
Q3 = movies_df['BoxOffice (Crores)'].quantile(0.75)
IQR = Q3 - Q1
```

```python
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = movies_df[movies_df['BoxOffice (Crores)'] > upper_bound]
outliers = movies_df[['Title', 'BoxOffice (Crores)', 'Industry']]

outliers.sort_values(by='BoxOffice (Crores)', ascending=False).head(15)
```
