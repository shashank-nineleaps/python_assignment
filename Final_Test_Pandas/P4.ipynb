```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
movies_data = './Datasets/movies_data_cleaned/combined_dataset.csv'
```

```python
movies_df = pd.read_csv(movies_data)
```

```python
movies_df = movies_df.drop_duplicates(subset=['Title'], keep='first')
```

```python
movies_df = movies_df.drop_duplicates(subset=['MovieID'], keep='first')
```

```python
movies_df = pd.read_csv(movies_data)
```

```python
genre_rating = movies_df.groupby('Genre')['Rating'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=genre_rating, x='Rating', y='Genre', hue = 'Genre', palette='cubehelix')
plt.title('Average IMDb Rating by Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies_df, x='Budget (Crores)', y='BoxOffice (Crores)', hue='Industry', palette='Set1')
sns.regplot(data=movies_df, x='Budget (Crores)', y='BoxOffice (Crores)', scatter=False, color='black')
plt.title('Budget vs BoxOffice Collection')
plt.xlabel('Budget (Crores)')
plt.ylabel('BoxOffice (Crores)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies_df, x='Duration (minutes)', y='Rating', hue='Industry', palette='Dark2')
plt.title('Duration vs IMDb Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
lang_boxoffice = movies_df.groupby('Language')['BoxOffice (Crores)'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=lang_boxoffice, x='BoxOffice (Crores)', y='Language', hue='Language', palette='rocket')
plt.title('Average BoxOffice by Language')
plt.xlabel('Average BoxOffice (Crores)')
plt.ylabel('Language')
plt.tight_layout()
plt.show()
```
